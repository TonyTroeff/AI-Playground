# Named Entity Recognition (Local Models)

Use this notebook to experiment with Hugging Face named entity recognition (`ner`) models while storing model files **inside this project workspace**.

```python
import os
from dotenv import load_dotenv

os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
os.environ["HF_HUB_CACHE"] = "./huggingface"
_ = load_dotenv(dotenv_path=".env.local")
```

```python
import torch

from transformers import TokenClassificationPipeline, pipeline
from typing import List
```

```python
experiments: List[str] = [
    "I'm in Plovdiv on Friday and want to hear live jazz without wasting the whole evening on travel. Check the current mini tour for \"Vassil Petrov & JP3\", figure out the relevant venue, and estimate the drive time.",
    "Satya Nadella will meet Ursula von der Leyen in Brussels on April 12 to discuss AI regulation.",
    "Apple is opening a new research office in Zurich after acquiring a startup founded by Maria Ivanova.",
    "Send the signed contract from Acme Corp to Dr. James Wilson at Mayo Clinic in Rochester, Minnesota.",
    "Book a table at Nobu London for two people next Tuesday at 7 pm under the name Elena Petrova.",
    "The United Nations scheduled a climate summit in Nairobi for September 2026 with delegates from Brazil, Japan, and Canada.",
    "Jean Dupont flew from Paris to New York to present a partnership between Airbus and IBM.",
]
```

```python
def get_local_named_entity_recognizer(model_id: str) -> TokenClassificationPipeline:
    device = "cuda" if torch.cuda.is_available() else "cpu"

    print(f"Loading model \"{model_id}\" on {device}")
    return pipeline(task="token-classification", model=model_id, device=device)
```

```python
def run_named_entity_recognition_experiment(model_id: str):
    named_entity_recognizer = get_local_named_entity_recognizer(model_id)

    print(f"{'=' * 10} [{model_id}] {'=' * 10}")

    for i in range(len(experiments)):
        print()

        print(experiments[i])
        result = named_entity_recognizer(experiments[i], aggregation_strategy="simple")

        for entity in result:
            print(f"  - {entity["entity_group"]} ({entity["score"]:.4%}): \"{entity["word"]}\"")
```

```python
run_named_entity_recognition_experiment("dbmdz/bert-large-cased-finetuned-conll03-english")
```
